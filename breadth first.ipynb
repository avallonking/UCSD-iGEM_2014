```python
graph = {'A': set(['B', 'C']),
         'B': set(['A','D']),
         'C': set(['F']),
         'D': set(['E']),
         'E': set(['F','L']),
         'F': set(['G']),
         'G': set(['H','I','J']),
         'H': set(['J']),
         'I': set(['H']),
         'J': set(['M']),
         'K': set(['G','H']),
         'L': set(['K','O']),
         'M': set(['G', 'N']),
         'N': set(['L']),
         'O': set(['P']),
         'P': set(['O','Q']),
         'Q': set(['R']),
         'R': set(['S']),
         'S': set(['E','T']),
         'T': set(['U']),
         'U': set(['V']),
         'V': set(['S','W','X']),
         'W': set(['B']),
         'X': set(['R','Y']),
         'Y': set(['Z']),
         'Z': set([])}

def bfs_paths(graph, start, goal):
    queue = [(start, [start])]
    while queue:
        (vertex, path) = queue.pop(0)
        for next in graph[vertex] - set(path):
            if next == goal:
                yield path + [next]
            else:
                queue.append((next, path + [next]))

list(bfs_paths(graph, 'A', 'E')) 
```
