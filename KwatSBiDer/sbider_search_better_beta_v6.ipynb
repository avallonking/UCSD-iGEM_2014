```python
def remove_duplicates_within_list(seq):
    seen = set()
    seen_add = seen.add
    return [ x for x in seq if not (x in seen or seen_add(x))]
```

```python
def list_is_type(lst, typ):
    """
    Check if all elements in a list is the same specified type.
    """
    
    if type(lst) != list:
        raise TypeError("list_is_type(lst, typ): lst is not a list")
    elif len(lst) <= 0:
        raise ValueError("list_is_type(lst, typ): lst is empty")
    return all(isinstance(x,typ) for x in lst)
```

```python
def remove_duplicated_lists_within_a_list_of_lists(list_of_lists):
    """
    Removes any duplicated lists, which contain same elements in same order, within a list of lists.
    """
    
    if type(list_of_lists) == list and len(list_of_lists) > 0 and list_is_type(list_of_lists, list) == True:
        list_of_lists.sort()
        trimmed = list(list_of_lists for list_of_lists,_ in itertools.groupby(list_of_lists))
        return trimmed
    else:
        raise TypeError("remove_duplicated_lists_within_a_list_of_lists(list_of_lists): list_of_lists should be in the form: [[],[],... ]")
```

```python
def uniquely_merge_list_of_lists(list_of_lists):
    """
    Merge unique elements from lists within a list.
    
    Argument:
        list_of_lists - a list that contains multiple lists.
    
    Return:
        a new list that contains unique elements from all lists within a list of lists.
    """   
    
    if type(list_of_lists) == list and len(list_of_lists) > 0:
        
        if list_is_type(list_of_lists, list) == True:
            remove_duplicated_lists_within_a_list_of_lists(list_of_lists)
            merged_list = list(list_of_lists[0])
            for a_list in list_of_lists[1::]:
                    for e in a_list:
                        if e not in merged_list:
                            merged_list.append(e)
            return merged_list
        else:
            return list(list_of_lists)
    else:
        raise TypeError("uniquely_merge_list_of_lists(list_of_lists): list_of_list must be a non-empty list")
```

```python
def uniquely_merge_multi_dimentional_list_of_lists(multi_dimentional_list_of_lists):
    final_merged_list = uniquely_merge_list_of_lists(multi_dimentional_list_of_lists)
    if type(final_merged_list) == list and len(final_merged_list) > 0 and list_is_type(final_merged_list, list) == True:
        return uniquely_merge_multi_dimentional_list_of_lists(final_merged_list)
    else:
        return final_merged_list
```

```python
def elements_match(list_of_lists, lst):
    return set(lst).issubset(uniquely_merge_multi_dimentional_list_of_lists(list_of_lists))
```

```python
def get_matching_list(list_of_lists, lst):
    """
    Get any list that match a specified list.
    """
    
    if type(lst) != list:
        raise TypeError("get_matching_list(list_of_lists, lst): lst is not a list object")    
    elif type(list_of_lists) == list and len(list_of_lists) > 0 and list_is_type(list_of_lists, list) == True:
        matching_list = []
        for a_list in list_of_lists:
            if all([x in lst for x in a_list]) == True:
                matching_list.append(a_list)
        return matching_list
    else:
        raise TypeError("get_matching_list(list_of_lists, lst): list_of_lists should be in the form: [[],[],... ]")
```

```python
def match_any_list(list_of_lists, lst):
    """
    Check if a list matches any of lists within a list.
    """
            
    if get_matching_list(list_of_lists, lst) == []:
        return False
    else:
        return True
```

```python
def sbider_path(input_dictionary,
                output_dictionary,
                input_species_list,
                output_species_list,
                path_queue,
                final_operon_path_list,
                final_species_path_list,
                memory_operon,
                memory_species,
                species_path_queue,
                memory_path_queue):
    """Construct path"""
    
    print "#" * 99
    print "********* sbider_path: path_queue:", path_queue, len(path_queue)
    print "********* sbider_path: final_operon_path_list:", final_operon_path_list
    print "********* sbider_path: final_species_path_list:", final_species_path_list
    print "********* sbider_path: memory_operon:", memory_operon
    print "********* sbider_path: memory_species:", memory_species
    print "********* sbider_path: species_path_queue:", species_path_queue
    print "#" * 99
        
    while len(path_queue) != 0:
        print "\t ######### sbider_path: <per pop> path_queue:", path_queue, len(path_queue)
        print "\t ######### sbider_path: <per pop> memory_operon:", memory_operon 
        print "\t ######### sbider_path: <per pop> memory_species:", memory_species
        print "\t ######### sbider_path: <per pop> species_path_queue:", species_path_queue, len(species_path_queue)
        
        (previously_visited_operon_list, just_previously_produced_species_list) = path_queue.pop(0)
        previously_visited_species_list = species_path_queue.pop(0)
        
        print "\n\t sbider_path: searching for operons that can be activated by:", just_previously_produced_species_list
        print "\t sbider_path: operons yet to be visited:", set(input_dictionary.keys()) - set(previously_visited_operon_list)
        for an_operon in set(input_dictionary.keys()) - set(previously_visited_operon_list):
            
            activation_requirement = input_dictionary[an_operon]
            
            print "\n\t\t sbider_path: potential operon & its input species:", an_operon, " & ", activation_requirement
            print "\t\t sbider_path: (ACTIVATE?)Does species", just_previously_produced_species_list, "match an_operon's input species:", activation_requirement,"?"
            if match_any_list(activation_requirement, just_previously_produced_species_list) == True:

                print "\n\t\t\t sbider_path: !!!ACTIVATION!!!"
                print "\t\t\t sbider_path: <OPERON>"
                print "\t\t\t sbider_path: (<before>) previously_visited_operon_list:", previously_visited_operon_list
                print "\t\t\t sbider_path: (<change>) an_operon:", an_operon
                visited_operon_list = previously_visited_operon_list + [an_operon]
                print "\t\t\tsbider_path: (<after>) visited_operon_list:", visited_operon_list

                print "\t\t\t sbider_path: <SPECIES>"
                print "\t\t\t sbider_path: (<before>) previously_visited_species_list:", previously_visited_species_list
                just_produced_species_list = output_dictionary[an_operon]
                print "\t\t\t sbider_path: (<change>) just produced species by", "operon (ope_id", an_operon, "):", just_produced_species_list                
                visited_species_list = previously_visited_species_list + uniquely_merge_multi_dimentional_list_of_lists(just_produced_species_list)
                visited_species_list = remove_duplicates_within_list(visited_species_list)
                print "\t\t\t sbider_path: (<after>) visited_species_list:", visited_species_list
                
                print "\t\t\t sbider_path: Can", output_species_list, "be found in", just_produced_species_list,"?"
                
                if match_any_list(just_produced_species_list, output_species_list):
                    print "\n\t\t\t\t sbider_path: !!!OUTPUT SPECIES FOUND!!! \n"
                    final_operon_path_list.append(visited_operon_list)
                    final_species_path_list.append(list(visited_species_list))
                    print "\n\t\t\t\t sbider_path: final_operon_path_list:", final_operon_path_list
                    print "\t\t\t\t sbider_path: final_species_path_list:", final_species_path_list
                else:        
                    #if an_operon not in memory_operon:
                    print "\t\t\t\t sbider_path: operon(ope_id:", an_operon, ") does not produce output species, but", just_produced_species_list, "-> path_queue"
                    path_queue.append( (visited_operon_list,uniquely_merge_multi_dimentional_list_of_lists(just_produced_species_list)) )
                    memory_species.extend(visited_species_list)
                    memory_species = remove_duplicates_within_list(memory_species)
                    species_path_queue.append(visited_species_list)
                          
                    # [ [[ope_path1],[spe_path1]], [[ope_path2],[spe_path2]], ...]
                    memory_path_queue.append( [visited_operon_list,uniquely_merge_multi_dimentional_list_of_lists(just_produced_species_list)] )
                    print " \n\t\t\t\t sbider_path: memory_path_queue:", memory_path_queue
                    print "\n"
                        
                memory_operon.append(an_operon)
                memory_operon = remove_duplicates_within_list(memory_operon)
                
                print "\n\t\t\t\t\t ######### sbider_path: <per operon> path_queue:", path_queue, len(path_queue)
                print "\t\t\t\t\t ######### sbider_path: <per operon> memory_operon:", memory_operon 
                print "\t\t\t\t\t ######### sbider_path: <per operon> memory_species:", memory_species
                print "\t\t\t\t\t ######### sbider_path: <per operon> species_path_queue:", species_path_queue, len(species_path_queue)
                print "\n\n"
    
    # if there is species produced, run sbider_path_aggr(...).
    print "##### sbider_path: ENTERING- sbider_path_aggr #####"
    sbider_path_aggr(input_dictionary,
                     output_dictionary,
                     input_species_list,
                     output_species_list,
                     path_queue,
                     final_operon_path_list,
                     final_species_path_list,
                     memory_operon,
                     memory_species,
                     species_path_queue,
                     memory_path_queue)
    print "##### sbider_path: -EXITING sbider_path_aggr #####"
    print
    print
    if len(final_operon_path_list) > 0:
         final_operon_path_list = remove_duplicated_lists_within_a_list_of_lists(final_operon_path_list)
    if len(final_species_path_list) > 0:
         final_species_path_list = remove_duplicated_lists_within_a_list_of_lists(final_species_path_list)
    return final_operon_path_list, final_species_path_list
```
