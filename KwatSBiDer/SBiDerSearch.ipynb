```python
def r_index(sequence, element):
    '''find the last index of an element in a sequence'''

    for i, e in enumerate(reversed(sequence)):
        if element == e:
            return len(sequence) - 1 - i
    else:
            raise ValueError("r_index(sequence, element): element not in the sequence")
```

```python
def removeParentheses(sequence):
    '''remove the outer most parentheses "()" 
    and returns the token afte the ")"'''

    #print "removeParentheses(sequence): sequence: ", sequence
    
    firstOpenerIndexAssigned = False
    started = False
    counter = 0
    
    for idx, e in enumerate(sequence):
        if e == '(':
            if started == False:
                started = True
            counter = counter + 1
        elif e == ')':
            if started == False:
                raise ValueError("removeParentheses(sequence): ')' without '('")
            counter = counter - 1
            
        if started == True:
            if firstOpenerIndexAssigned == False:
                firstOpenerIndex = idx
                firstOpenerIndexAssigned = True
            if counter == 0:
                sequence.pop(idx)
                if idx < len(sequence):
                    element_after_last_closer = sequence[idx]
                else:
                    element_after_last_closer = None
                sequence.pop(firstOpenerIndex)
                return element_after_last_closer
                
    return None
```

```python
def splitBy(sequence, element):
    '''splits a sequence by a given element and stores elements
    before and after the element into a dictionary'''
    
    element_index = sequence.index(element)
    
    sequence_before_element = sequence[:element_index:1]
    sequence_after_element = sequence[element_index + 1::1]
    
    return {0: sequence_before_element, 1: sequence_after_element}
```

```python
def g0(tokens):
    '''g0:= g1 > g1'''

    #print "g0(tokens): tokens: ", tokens
    
    if '>' not in tokens:
        raise ValueError("g(tokens): no output")
        
    else:
        input_output_dictionary = splitBy(tokens, '>')
        
    return gOutput(g1(input_output_dictionary[0]), g1(input_output_dictionary[1]))
```

```python
def g1(tokens):
    '''g1:= g2 or g1 | g2 and g1 | g2'''

    #print "g1(tokens): tokens: ", tokens
    
    if len(tokens) > 1 and tokens[1] == 'or':
        "g2 or g1"
        
        #print "g0(tokens): detected 'or'"
        
        "splits tokens by the first occuring 'or' and stores the tokens before and after the 'or' in a dictionary"
        or_dictionary = splitBy(tokens, 'or')
        return gOr(g2(or_dictionary.get(0)), g1(or_dictionary.get(1)))
    elif len(tokens) > 1 and tokens[1] == 'and':
        "g2 and g1"
        
        #print "g1(tokens): detected 'and'"
        
        "splits tokens by the first occuring 'and' and stores the tokens before and after the 'and' in a dictionary"
        and_dictionary = splitBy(tokens, 'and')
        return gAnd(g2(and_dictionary.get(0)), g1(and_dictionary.get(1)))            
    else:
        "g2"
        "delegates to g2"
        return g2(tokens)
```

```python
def g2(tokens):
    '''g2:= (g1) or g1 | (g1) and g1 | (g1) | interactor''' 

    #print "g2(tokens): tokens: ", tokens
    
    if len(tokens) <= 0:
        raise ValueError("Invalid Syntax")
    elif tokens[0] == "(":
        "(g1) or g1 | (g1) and g1| (g1)"
        
        #print "g2(tokens): detected '('"
        
        "token after the last occuring ')'"
        token_after_last_closer = removeParentheses(tokens)
        if token_after_last_closer == 'or':    
            "splits tokens by the first occuring 'or' and stores the tokens before and after the 'or' in a dictionary"
            or_dictionary = splitBy(tokens, 'or')
            return gOr(g1(or_dictionary.get(0)), g1(or_dictionary.get(1)))
        elif token_after_last_closer == 'and':
            "splits tokens by the first occuring 'and' and stores the tokens before and after the 'and' in a dictionary"
            and_dictionary = splitBy(tokens, 'and')
            return gAnd(g1(and_dictionary.get(0)), g1(and_dictionary.get(1)))
        else:
            "delegates to interactor"
            return g1(tokens)
    else:
        "interactor"
        "delegates to interactor"
        return interactor(tokens)
```

```python
def interactor(tokens):
    '''interactor:= lacI | cI | luxR | ... | etc.'''

    #print "interactor(tokens): tokens: ", tokens
    
    "returns an interactor"
    return (tokens)
```

```python
def gOutput(tokens1, tokens2):
    '''>'''

    toBeReturned = {}
    for token1 in tokens1:
        for token2 in tokens2: 
            toBeReturned[token1] = (token2)
    return toBeReturned
```

```python
def gOr(tokens1, tokens2):
    '''or'''

    return tokens1 + tokens2
```

```python
def gAnd(tokens1, tokens2):
    '''and'''

    toBeReturned = []
    for token1 in tokens1:
        for token2 in tokens2:
            toBeReturned.append((token1,token2))
    return toBeReturned
```

```python
def parseInputLogic(userInput):
    '''parses a user input into atomized and equivalent logics'''
    
    #print "parseInputLogic(tokens): tokens: ", tokens
    
    tokens = userInput.split()
    return g0(tokens)
```

```python
def uniteLists(listOfLists):
    '''unites multiple lists'''
    
    unitedList = list(listOfLists[0])
    for aList in listOfLists[1::]:
        for e in aList:
            if e not in unitedList:
                unitedList.append(e)
        return unitedList
```

```python
def getOperonActivated(requirement, inpt):
    operonActivatedByAnd = []
    for aRequirement in requirement:
        if all([x in inpt for x in aRequirement]) == True:
            operonActivatedByAnd.append(aRequirement)
    
    if len(operonActivatedByAnd) > 0:
        return operonActivatedByAnd
    else:
        return None
```

```python
def isActivated(requirement, inpt):
    if getOperonActivated(requirement, inpt) == None:
        return False
    else:
        return True
```

```python
def getPath(inputDictionary, outputDictionary, inputSpeciesList, outputSpeciesList):
    pathQueue = [ ([],inputSpeciesList) ]    
    finalOperonPathList = []
    finalSpeciesPathList = []
    while pathQueue != []:
        
        #print "\ngetPath loop(...)"

        pathQueue, finalOperonPathList, finalSpeciesPathList = traverse(inputDictionary,
                                                                        outputDictionary,
                                                                        inputSpeciesList,
                                                                        outputSpeciesList,
                                                                        pathQueue,
                                                                        finalOperonPathList,
                                                                        finalSpeciesPathList)
    return finalOperonPathList, finalSpeciesPathList
```

```python
def traverse(inputDictionary,
             outputDictionary,
             inputSpeciesList,
             outputSpeciesList,
             pathQueue,
             finalOperonPathList,
             finalSpeciesPathList):
    
    (visitedOperonList, visitedSpeciesList) = pathQueue.pop(0)
    
    # visited species become input species
    anInputSpeciesList = list(visitedSpeciesList)

    #print "\tanInputSpeciesList: ",anInputSpeciesList

    # searches for operons activated by current input species
    for anOperon in set(inputDictionary.keys()) - set(visitedOperonList):
        
        #print "\t\tanOperon & potential output species: ",anOperon,"&",inputDictionary[anOperon]
        #print "\t\tcan",anInputSpeciesList,"activate",inputDictionary[anOperon],"???"
        
        if isActivated(inputDictionary[anOperon], anInputSpeciesList) == True:
            
            print "\t\t\tyes!"
            
            # activated operon and its unique outout species are stored
            aVisitedOperonList = list(visitedOperonList) + [anOperon]
            aVisitedSpeciesList = uniteLists([visitedSpeciesList, outputDictionary[anOperon]])

            #print "\t\t\taVisitedOperonList: ",aVisitedOperonList
            #print "\t\t\taVisitedSpeciesList: ",aVisitedSpeciesList
            #print "\t\t\toutputSpeciesList: ",outputSpeciesList
            
            # if output speices is found, include this operon and species path lists in the 
            # final operon and species paths
            if set(outputSpeciesList).issubset(aVisitedSpeciesList):
                
                #print "\t\t\t\toutput species found!"
                
                finalOperonPathList.append(aVisitedOperonList)
                finalSpeciesPathList.append(aVisitedSpeciesList)
                
            # If output species is not found, include this operon and species path lists in path queue
            else:
                
                #print "\t\t\t\toutput species not found..."
                
                pathQueue.append( (aVisitedOperonList,aVisitedSpeciesList) )
    
    print "\t\tfinal pathQueue: ",pathQueue
    
    return pathQueue, finalOperonPathList, finalSpeciesPathList
```

```python
'''make networkx graph
@param - pathList: list representation of path:
        [A, B, C, ...]
@return - networkx directed graph'''
def makeGraph(pathList):
    
    graph = nx.DiGraph()
    
    if len(pathList) < 2:
        graph.add_nodes_from(pathList)
    else:
        graph.add_path(pathList)
        
    return graph
```

```python
'''expand networkx graph
@param - graph: graph to be extended
@param - pathList: list representation of path'''
def expandGraph(graph, pathList):
    graph.add_path(pathList)
```

```python
'''draw and plot a graph
@param - graph: graph to be drawn and plotted'''
def drawPlot(graph):
    
    nx.draw(graph,
            node_size=1600,
            node_color='black',
            font_size=16,
            font_color='yellow',
            width=1,
            edge_color='grey',
            style='dotted',
            arrows=True)
    
    plt.show()
```
